```python
def get_links(collected_links):

    critic_urls = []
    audience_urls = []

    base_url = 'https://www.rottentomatoes.com/'

    for link in tqdm(collected_links):
        soup = read_cloud(link)
        # get critic review links
        try:
            critic_url = soup.find_all(class_ = 'link')[0]['href']
        except:
            critic_url = 'None'

        # get audience review links
        try:
            audience_url = soup.find_all(class_ = 'link')[1]['href']
        except:
            audience_url = 'None'

        # create valid url
        critics_url = base_url + critic_url
        audience_url = base_url + audience_url

        # create list of urls with appending individual urls to an empty list
        critic_urls.append(critics_url)
        audience_urls.append(audience_url)
    return critic_urls, audience_urls
```

```python
urls,audience_urls = get_links(drama_collected_links)
```

```python
# function to read page
def read_cloud(link):
    scraper = cloudscraper.create_scraper()
    response = scraper.get(link)
    return(BeautifulSoup(response.text, 'html'))


# function to get critic reviews
def get_critics(urls):

    reviews = [] # reviews
    names = [] # movie names 
    
    for url in tqdm(urls):
        soup = read_cloud(url) # read page
        critics = soup.find_all(class_='review-text') # find all text boxes for reviews
        review = [x.text for x in critics] # get text from reviews
        reviews.append(review) # append to list
        names.append(url.split('?')[0]) # append to list and create clean name format from the url
        critic_reviews = {'name': names, 'critic_reviews': reviews} # create dictionary
    return critic_reviews

# function to get audience reviews
def get_audience(audience_urls):

    reviews = [] # reviews
    names = [] # movie names
    
    for url in tqdm(audience_urls):
        soup = read_cloud(url)  # read page
        critics = soup.find_all(class_='audience-reviews__review js-review-text') # find all text boxes for reviews
        try:
            review = [x.text.strip() for x in critics] # get text from reviews
        except:
            review = 'None'
        reviews.append(review) # append to list
        names.append(url.split('?')[0]) # append to list and create clean name format from the url
        audience_reviews = {'name': names, 'audience_reviews': reviews} # create dictionary
    return audience_reviews
```

```python
critic_reviews = get_critics(urls)
audience_reviews = get_audience(urls)
```

```python
critic = pd.DataFrame(critic_reviews)
audience = pd.DataFrame(audience_reviews)
drama_df = critic.merge(audience, on='name')
```

```python
drama_df['critic_reviews'] = drama_df['critic_reviews'].apply(lambda x: ' '.join(x)) # flatten the list
drama_df['audience_reviews'] = drama_df['audience_reviews'].apply(lambda x: ' '.join(x)) # flatten the list
drama_df['name'] = drama_df['name'].apply(lambda x: x.split('/')[-2]) # get the name of the movie
drama_df['critic_reviews'].replace('', pd.NA, inplace=True)  # replace empty strings with NA
drama_df.dropna(subset=['critic_reviews'], inplace=True)  # drop rows with NA
drama_df['genre'] = 'Drama' # add genre
drama_df = drama_df[['genre','name','critic_reviews', 'audience_reviews']] # reorder columns
drama_df.reset_index(drop=True, inplace=True) # reset index
drama_df
```

# Sentiment Analysis

```python
import pprint

import boto3

pp = pprint.PrettyPrinter(indent=2)
# %%
# Language Detection
comprehend = boto3.client(service_name="comprehend", region_name="eu-west-1")
```

```python
# since there are there are duplicated films (they can be the same film for different genres)
# we will drop the duplicates before running sentiment analysis to optimize costs and then merge the results back to the original dataframe
```

```python
sent_df = df.drop_duplicates(subset=['name'])[['name', 'critic_reviews', 'audience_reviews']] # drop duplicates
sent_df.reset_index(drop=True, inplace=True) # reset index
sent_df
```

```python
# function to run sentiment analysis
def run_sentiment_analysis(text):
    sentiment_result = comprehend.detect_sentiment(Text=text, LanguageCode="en")
    return sentiment_result
```

```python
total_length = sum(len(review) for review in sent_df[['critic_reviews', 'audience_reviews']].values.flatten())
print(f"Running the sentiment analysis with {total_length} characters")
```

```python
# Apply sentiment analysis and extract required information
sent_df['critic_sentiment_result'] = sent_df['critic_reviews'].apply(run_sentiment_analysis)
sent_df['audience_sentiment_result'] = sent_df['audience_reviews'].apply(run_sentiment_analysis)

# Extract positive, negative, neutral, mixed, and overall sentiment
sent_df['critic_postive'] = sent_df['critic_sentiment_result'].apply(lambda x: x['SentimentScore']['Positive'])
sent_df['critic_negative'] = sent_df['critic_sentiment_result'].apply(lambda x: x['SentimentScore']['Negative'])
sent_df['critic_neutral'] = sent_df['critic_sentiment_result'].apply(lambda x: x['SentimentScore']['Neutral'])
sent_df['critic_mixed'] = sent_df['critic_sentiment_result'].apply(lambda x: x['SentimentScore']['Mixed'])
sent_df['critic_sentiment'] = sent_df['critic_sentiment_result'].apply(lambda x: x['Sentiment'])

sent_df['audience_postive'] = sent_df['audience_sentiment_result'].apply(lambda x: x['SentimentScore']['Positive'])
sent_df['audience_negative'] = sent_df['audience_sentiment_result'].apply(lambda x: x['SentimentScore']['Negative'])
sent_df['audience_neutral'] = sent_df['audience_sentiment_result'].apply(lambda x: x['SentimentScore']['Neutral'])
sent_df['audience_mixed'] = sent_df['audience_sentiment_result'].apply(lambda x: x['SentimentScore']['Mixed'])
sent_df['audience_sentiment'] = sent_df['audience_sentiment_result'].apply(lambda x: x['Sentiment'])
```

```python
df = df[['genre','name','critic_score','audience_score']].merge(sent_df, on='name') # merge the sentiment analysis results back to the original dataframe and only keep unique columns
df
```

```python
# export to csv
df.to_csv('reviews2.csv', index=False)
```

# Get poster images

```python
def get_images(collected_links):
    for link in tqdm(collected_links):
        
        soup = read_cloud(link)
        # select the image link class
        img_link = soup.find(class_ = 'thumbnail')
        # select the link for the image
        image_src = img_link.select_one('rt-img[src]')['src']
        # specify file path
        file_path = f"data/images/{link.split('/')[-1]}_poster.jpg"
        # Download the image
        urllib.request.urlretrieve(image_src, file_path)
    print(f"Downloaded {len(collected_links)} images")

```
