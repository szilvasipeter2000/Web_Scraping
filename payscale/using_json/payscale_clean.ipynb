```python
# import libraries
import pandas as pd
import ast
```

```python
# readinig the raw data
df_location = pd.read_csv('https://raw.githubusercontent.com/szilvasipeter2000/Web_Scraping/main/payscale/using_json/data/raw/data_location.csv')
df_employer = pd.read_csv('https://raw.githubusercontent.com/szilvasipeter2000/Web_Scraping/main/payscale/using_json/data/raw/data_employer.csv')
df_skills = pd.read_csv('https://raw.githubusercontent.com/szilvasipeter2000/Web_Scraping/main/payscale/using_json/data/raw/data_skill.csv')
df_experience = pd.read_csv('https://raw.githubusercontent.com/szilvasipeter2000/Web_Scraping/main/payscale/using_json/data/raw/data_experience.csv')
df_summary = pd.read_csv('https://raw.githubusercontent.com/szilvasipeter2000/Web_Scraping/main/payscale/using_json/data/raw/data_summary.csv')
```

```python
# function to be used on all dimensions

# create function to extract job title from url
def extract_job_title(df,url):
    pattern = r'Job=(.*?)/Salary'
    df['job_title'] = df[url].str.extract(pattern)
    return df

# crate function to clean job title column as some titles have random characters in them
def clean_job_title(df):
    df['job_title'] = df['job_title'].apply(lambda x: x.replace('%2C','').replace('%2F','').replace('%26','').replace('%23','').replace('%2B%2B','') if isinstance(x, str) else x)
    return df

# create function to extract median salary from range column
def get_median_salary(df,range):
    df['median_salary'] = df[range].apply(lambda x: list(ast.literal_eval(x).values())[0] if x else None) # ast.literal_eval() to converts string to dictionary so that way i can just get the values as a list --> then select [0] element to get it as a float
    return df

```

```python
# df_location specific functions
def separate_city_state(df):
    df['city'] = df['name'].apply(lambda x: x.split(',')[0])
    df['state'] = df['name'].apply(lambda x: x.split(',')[1])
    df = df.drop(columns=['name']) # drop the 'name' column as we don't need it anymore
    return df
```

```python
df_location = extract_job_title(df_location, 'url')
df_location = get_median_salary(df_location,'range')
df_location = separate_city_state(df_location)
df_location = clean_job_title(df_location)
df_location['profile_count'] = df_location['profileCount']
df_location['state'] = df_location['state'].apply(lambda x: x.strip()) # remove whitespace from state column
df_location = df_location[['state','city','job_title','median_salary','profile_count']] # reorder columns and drop unnecessary columns
df_location
```

```python
# export it 
df_location.to_csv('clean_data_location.csv',index=False)
```

```python
#categorize them
def categorize_exp(df):

    experience_levels = {
    'Less than 1 year': 'Entry Level',
    '1-4 years': 'Junior Level',
    '5-9 years': 'Mid Level',
    '10-19 years': 'Senior Level',
    '20 years or more': 'Executive Level'
    }

    df['level']= df['name'].map(experience_levels)
    return df
```

```python
df_experience = extract_job_title(df_experience, 'url')
df_experience = clean_job_title(df_experience)
df_experience = get_median_salary(df_experience,'range')
df_experience = categorize_exp(df_experience)
df_experience[['profile_count','experience']] = df_experience[['profileCount','name']] # rename columns
df_experience = df_experience[['level','experience','job_title','median_salary','profile_count']] # reorder columns
df_experience
```

```python
# export it 
df_experience.to_csv('clean_data_experience.csv',index=False)
```

```python
def clean_summary_table(df):
    # split category to good format
    df['category'] = df['category'].apply(lambda x: x.split(' - ')[1] if isinstance(x, str) and ' - ' in x else x)

    #first extract the hourly and annual wages
    pattern = r'\$([\d,]+)'
    try:
        df['annual_wage'] = df['summary_salary'].str.extract(pattern).replace(',','',regex=True).astype(float)
    except:
        df['annual_wage'] = None
    try:
        df['hourly_wage'] = df['summary_hourly'].str.extract(pattern).replace(',','',regex=True).astype(float)
    except:
        df['hourly_wage'] = None

    # drop the unused columns
    df = df.drop(columns=['summary_hourly','summary_salary'])

    # reorder columns
    df = df[['job_title','category','description','annual_wage','hourly_wage','currency','country','rating','rating_count','gender_responses','male_count','female_count',]]
    return df
```

```python
df_summary = clean_summary_table(df_summary)
df_summary
```

```python
# export it 
df_summary.to_csv('clean_data_summary.csv',index=False)
```
